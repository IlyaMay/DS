### –†–∞—Å—á—ë—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']

recovery = (C * (F - T)) / (F * (C - T)) * 100
print(recovery)
```

–í—ã—á–∏—Å–ª–∏–º MAE –º–µ–∂–¥—É –Ω–∞—à–∏–º–∏ —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery

```python
mae_rec = mean_absolute_error(data_train['rougher.output.recovery'], recovery)
print('MAE =', mae_rec)
```

### –ò—Å—Å–ª–µ–¥—É–π—Ç–µ —Å—É–º–º–∞—Ä–Ω—É—é –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—é –≤—Å–µ—Ö –≤–µ—â–µ—Å—Ç–≤ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —Å—Ç–∞–¥–∏—è—Ö: –≤ —Å—ã—Ä—å–µ, –≤ —á–µ—Ä–Ω–æ–≤–æ–º –∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–º –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–∞—Ö.

```python
def sum_concentarution(row):
    list_of_steps = ["rougher","primary_cleaner","final"]
    input_feed_au = row["rougher.input.feed_au"]
    input_feed_ag = row["rougher.input.feed_ag"]
    input_feed_pb = row["rougher.input.feed_pb"]
    out_rougher = []
    out_primary_cleaner = []
    out_final = []
    list_of_arrays =[out_rougher,out_primary_cleaner,out_final]
    for step, array in zip(list_of_steps,list_of_arrays):
        array.append(row[step+".output.concentrate_au"])
        array.append(row[step+".output.concentrate_ag"])
        array.append(row[step+".output.concentrate_pb"])
    sum_bef_steps = input_feed_ag + input_feed_au + input_feed_pb 
    sum_rougher = sum(out_rougher)
    sum_primary = sum(out_primary_cleaner)
    sum_final = sum(out_final)
    return pd.Series([sum_bef_steps, sum_rougher, sum_primary, sum_final])
```

```python
sum_val = data_train.apply(sum_concentarution, axis = 1)
sum_val.columns = ["before_rougher","rougher","primary_cleaner","final"]
sum_val
```

```python
metr = [
    'before_rougher', 'rougher',
    'primary_cleaner','final']

for metric in metr:
    sum_val[metric].plot(
            title = '–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–æ–≤ –≥—Ä–∞–Ω—É–ª —Å—ã—Ä—å—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö –¥–ª—è {}'.format(metric),
            kind='hist', 
            bins=100, 
            grid=True, 
            figsize=(15, 6),
            alpha=0.5,
            label=label,
        ).set(xlabel='–°—É–º–º–∞ –¥–æ–ª–µ–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤', ylabel = '–ß–∞—Å—Ç–æ—Ç–∞')

    plt.show()
```

```python
def calculate_smape(target_actual, target_predictions):
    num = np.abs(target_actual - target_predictions)
    den = (np.abs(target_actual) + np.abs(target_predictions)) / 2 
    
    return 1 / len(target_actual) * np.sum(num / den) * 100 
    
    
def calculate_final_smape(target_actual, target_predictions):
    target_actual_rougher = target_actual['rougher.output.recovery']
    target_actual_final = target_actual['final.output.recovery']
    
    target_predictions_rougher = target_predictions[:, 0]
    target_predictions_final = target_predictions[:, 1]
    
    return 0.25 * calculate_smape(target_actual_rougher, target_predictions_rougher) \
        + 0.75 * calculate_smape(target_actual_rougher, target_predictions_rougher)
```
